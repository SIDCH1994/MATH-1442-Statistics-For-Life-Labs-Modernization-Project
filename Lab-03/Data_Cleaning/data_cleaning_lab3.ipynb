# Dataset for Lab-3 is PAHSSnr.csv

PASenrios.csv file is extracted from Lock5Data library using R Studio

```python
import pandas as pd
```

```python
# Step-1: Importing the raw dataset.
# --------------------------------------------------
PAHSSnr = pd.read_csv(r'..\Data_Extraction\PASeniors.csv')
print(PAHSSnr.head())
```

```python
# Step-2: Remove variables 'Model' and 'Location'
# --------------------------------------------------
columns_to_drop = ['Superpower', 'Preference', 'SchoolPressure', 'Activity', 'Music', 'Sleep2']
PAHSSnr_1 = PAHSSnr.drop(columns=columns_to_drop)
print(f"\nColumns '{', '.join(columns_to_drop)}' removed.")
print(PAHSSnr_1.head())
```

```python
# Step-3: Removing rows with Null values
# --------------------------------------------------
initial_rows = PAHSSnr_1.shape[0]
PAHSSnr_1.dropna(inplace=True)
rows_removed = initial_rows - PAHSSnr_1.shape[0]
print(f"\nRemoved {rows_removed} rows with null values.")
```

```python
# Step-4: Renaming the column names
# --------------------------------------------------
rename_map = {'Armspan': 'Arm_span',
              'GetToSchool': 'Get_to_School',
              'TravelTime': 'Travel_Time',
              'ReactionTime': 'Reaction_Time',
              'MemoryScore': 'Memory_Score',
              'BirthMonth': 'Birth_Month',
              'FavFood': 'Fav_Food',
              'FavSubject': 'Fav_Sub',
              'Sleep1': 'Sleep_Time',
              'Communicate': 'Communication',
              'TextsSent': 'Txt_Sent',
              'HangHours': 'Hang_Hrs',
              'HWHours': 'HW_Hrs',
              'SportsHours': 'Sport_Hrs',
              'VideoGameHours': 'VG_Hrs',
              'ComputerHours': 'Comp_Hrs',
              'TVHours': 'TV_Hrs',
              'WorkHours': 'Work_Hrs'}

PAHSSnr_2 = PAHSSnr_1.rename(columns=rename_map)

PAHSSnr_2.head()
```

```python
# Step 5: Making the column names uppercase letters.
# --------------------------------------------------
PAHSSnr_2.columns = [col.upper() for col in PAHSSnr_2.columns]
PAHSSnr_2
```

```python
# Step-6: Exporting the file
# --------------------------------------------------
output_filename = 'PAHSSnr.csv'
PAHSSnr_2.to_csv(output_filename, index=False)
print(f"\nCleaned dataset exported to '{output_filename}' successfully.")
```
