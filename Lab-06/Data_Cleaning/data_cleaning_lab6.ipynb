# Dataset for Lab-6 is MLB_Raw.csv

Raw dataset file MLB_Raw.csv is extracted from the openintro library in RStudio

```python
# Step-1: Importing the raw dataset.
# --------------------------------------------------
import pandas as pd
MLB = pd.read_csv(r'..\Data_Extraction\MLB_Raw.csv')
```

```python
# Step-2: Removing unwanted variables
# --------------------------------------------------
cols_to_remove = ['rank', 'division_id', 'home_games', 'division_winner', 'wild_card_winner', 'world_series_winner', 
                  'league_winner', 'stolen_bases', 'caught_stealing', 'batters_hit_by_pitch', 'sacrifice_flies',
                  'shutouts', 'saves', 'home_attendance']

MLB_1 = MLB.drop(columns=cols_to_remove)
MLB_1.head()
```

```python
# Step-3: Renaming the column names
# --------------------------------------------------
rename_map = {'year': 'YEAR',
              'league_id': 'LG_ID',
              'games_played': 'GMS_PLYD',
              'wins': 'WINS',
              'losses': 'LOSSES',
              'runs_scored': 'RNS_SCRD',
              'at_bats': 'AT_BATS',
              'hits': 'HITS',
              'doubles': 'DOUBLES',
              'triples': 'TRIPLES',
              'homeruns': 'HM_RUNS',
              'walks': 'WALKS',
              'strikeouts_by_batters': 'SOBB',
              'opponents_runs_scored': 'ORS',
              'earned_runs_allowed': 'ERA',
              'earned_run_average': 'ER_AVG',
              'complete_games': 'COMP_GMS',
              'outs_pitches': 'OUTS_PCHS',
              'hits_allowed': 'HTS_ALWD',
              'homeruns_allowed': 'HRNS_ALWD',
              'walks_allowed': 'WKS_ALWD',
              'strikeouts_by_pitchers': 'SOBP',
              'errors': 'ERRS',
              'double_plays': 'DBL_PLYS',
              'fielding_percentage': 'FLDG_PERC',
              'team_name': 'TM_NAME',
              'ball_park': 'BL_PK'}

MLB_2 = MLB_1.rename(columns=rename_map)

MLB_2.head()
```

```python
# Step-4: Removing rows with NA or blank values.
# --------------------------------------------------
MLB_2.dropna(inplace=True)
```

```python
# Step-5: Exporting file as CSV to the current working directory
# --------------------------------------------------
output_file_name = 'MLB.csv'
MLB_2.to_csv(output_file_name, index=False)

print(f"Dataset successfully cleaned and saved to {output_file_name}")
```
