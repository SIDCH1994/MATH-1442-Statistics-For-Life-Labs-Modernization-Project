# Dataset for Lab-7 is Nutrition_Raw.csv

Raw dataset file Nutrition_Raw.csv is extracted from the openintro library in RStudio

```python
# Step-1: Importing the raw dataset.
# --------------------------------------------------
import pandas as pd
NS_data = pd.read_csv(r'..\Data_Extraction\Nutrition_Raw.csv')
NS_data
```

```python
# Step-2: Removing unwanted variables
# --------------------------------------------------
cols_to_remove = ['Alcohol', 'Vitamin']

NS_data_1 = NS_data.drop(columns=cols_to_remove)
NS_data_1.head()
```

```python
# Step-3: Renaming the column names
# --------------------------------------------------
rename_map = {'ID': 'ID',
              'Age': 'AGE',
              'Smoke': 'SMOKE',
              'Quatelet': 'QUATELET',
              'Calories': 'CALORIES',
              'Fat': 'FAT',
              'Fiber': 'FIBER',
              'Cholesterol': 'CHOLESTEROL',
              'BetaDiet': 'B_DIET',
              'RetinolDiet': 'RET_DIET',
              'BetaPlasma': 'B_PLASMA',
              'RetinolPlasma': 'RET_PLASMA',
              'Sex': 'SEX',
              'VitaminUse': 'VTMN_USE',
              'PriorSmoke': 'PR_SMOKE'}

NS_data_2 = NS_data_1.rename(columns=rename_map)

NS_data_2.head()
```

```python
# Step-4: Removing rows with NA or blank values.
# --------------------------------------------------
NS_data_2.dropna(inplace=True)
NS_data_2.head()
```

```python
# Step-5: Rename groups in the 'PR_SMOKE' column
# --------------------------------------------------
NS_data_2['PR_SMOKE'] = NS_data_2['PR_SMOKE'].map({1: 'Never', 2: 'Former', 3: 'Current'})
NS_data_2.head()
```

```python
# Step-6: Exporting file as CSV to the current working directory
# --------------------------------------------------
output_file_name = 'Nutrition.csv'
NS_data_2.to_csv(output_file_name, index=False)

print(f"Dataset successfully cleaned and saved to {output_file_name}")
```
