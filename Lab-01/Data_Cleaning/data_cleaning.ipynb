# Data Cleaning: NHANES_Raw.csv

NHANES_Raw.csv file is extracted from NHANES library using R Studio

```python
# Step-1: Importing the raw dataset.
# --------------------------------------------------
import pandas as pd
NR = pd.read_csv(r'..\Data_Extraction\NHANES_Raw.csv')
NR
```

```python
NR.columns
```

```python
# Step-2: Removing unwanted variables
# --------------------------------------------------
cols_to_remove = ['SurveyYr',  'AgeDecade', 'AgeMonths', 'Race3', 'HHIncome', 'HHIncomeMid',  'Length', 'HeadCirc', 
                  'BMICatUnder20yrs', 'BMI_WHO', 'BPSysAve', 'BPDiaAve', 'BPSys1', 'BPDia1', 'BPSys2', 'BPDia2',
                  'BPSys3', 'BPDia3', 'Testosterone', 'DirectChol',  'UrineFlow1', 'UrineVol2', 'UrineFlow2',
                  'DiabetesAge',  'DaysPhysHlthBad', 'DaysMentHlthBad', 'LittleInterest', 
                  'Depressed', 'nPregnancies', 'nBabies', 'Age1stBaby', 'SleepTrouble', 'PhysActive', 'PhysActiveDays',
                  'TVHrsDay', 'CompHrsDay', 'TVHrsDayChild', 'CompHrsDayChild', 'Alcohol12PlusYr', 'AlcoholDay', 
                  'AlcoholYear', 'SmokeNow', 'Smoke100', 'SmokeAge', 'Marijuana', 'AgeFirstMarij', 
                  'RegularMarij', 'AgeRegMarij', 'HardDrugs', 'SexEver', 'SexAge', 'SexNumPartnLife', 'SexNumPartYear', 
                  'SameSex', 'SexOrientation', 'PregnantNow']

NR_1 = NR.drop(columns=cols_to_remove)
NR_1.head()
```

```python
NR_1.shape
```

```python
# Step 3: Renaming the variable names
# ---------------------------------------------------
rename_map = {'ID': 'ID',
                         'Gender': 'GENDER',
                         'Age': 'AGE',
                         'Race1': 'RACE',
                         'Education': 'EDUCATION',
                         'MaritalStatus': 'MAR_ST',
                         'Poverty': 'POVERTY',
                         'HomeRooms': 'HOME_ROOMS',
                         'HomeOwn': 'HOME_TYPE',
                         'Work': 'WORK',
                         'Weight': 'WEIGHT',
                         'Height': 'HEIGHT',
                         'BMI': 'BMI',
                         'Pulse': 'PULSE',
                         'TotChol': 'TOT_CHOL',
                         'UrineVol1': 'URINE_VOL',
                         'Diabetes': 'DIABETES',
                         'HealthGen': 'HEALTH',
                         'SleepHrsNight': 'SLEEP_HRS',
                         'Smoke100n': 'SMOKE'}

NR_2 = NR_1.rename(columns=rename_map)

NR_2.head()
```

```python
NR_2.shape
```

```python
# Step-4: Removing rows with NA or blank values.
# --------------------------------------------------
NR_2.dropna(inplace=True)
NR_2.head()
```

```python
NR_2.shape
```

```python
# Step 5: Removing the duplicate rows
# ---------------------------------------------------
print(f"\nNumber of rows before removing duplicates: {len(NR_2)}")
NR_2.drop_duplicates(inplace=True)
print(f"Number of rows after removing duplicates: {len(NR_2)}")
```

```python
NR_2.head()
```

```python
# Step 6: Exporting the cleaned dataset to a CSV file
# ---------------------------------------------------
output_filename = 'NHANES.csv'
NR_2.to_csv(output_filename, index=False)
print(f"\nCleaned dataset exported to '{output_filename}' in the current working directory.")
```
