# Dataset for Lab-8 is SStudy_Raw.csv

Raw dataset file SStudy_Raw.csv is extracted from the Lock5Data library in RStudio

```python
# Step-1: Importing the raw dataset.
# --------------------------------------------------
import pandas as pd
SS = pd.read_csv(r'..\Data_Extraction\SStudy_Raw.csv')
SS
```

```python
# Step-2: Removing unwanted variables
# --------------------------------------------------
cols_to_remove = ['ClassesMissed', 'NumEarlyClass', 'AllNighter', 'EarlyClass']

SS_1 = SS.drop(columns=cols_to_remove)
SS_1.head()
```

```python
# Step-3: Renaming the column names
# --------------------------------------------------
rename_map = {'Gender': 'GENDER',
              'ClassYear': 'CL_YEAR',
              'LarkOwl': 'LNO',
              'GPA': 'GPA',
              'CognitionZscore': 'COG_ZSCORE',
              'PoorSleepQuality': 'PS_QUAL',
              'DepressionScore': 'DEP_SCORE',
              'AnxietyScore': 'ANX_SCORE',
              'StressScore': 'STR_SCORE',
              'DepressionStatus': 'DEP_STATUS',
              'AnxietyStatus': 'ANX_STATUS',
              'Stress': 'STRESS',
              'DASScore': 'DAS_SCORE',
              'Happiness': 'HAPPINESS',
              'AlcoholUse': 'ALCO_USE',
              'Drinks': 'DRINKS',
              'WeekdayBed': 'WD_BED',
              'WeekdayRise': 'WD_RISE',
              'WeekdaySleep': 'WD_SLEEP',
              'WeekendBed': 'WE_BED',
              'WeekdendRise': 'WE_RISE',
              'WeekendSleep': 'WE_SLEEP',
              'AverageSleep': 'AVG_SLEEP'}

SS_2 = SS_1.rename(columns=rename_map)

SS_2.head()
```

```python
# Step-4: Removing rows with NA or blank values.
# --------------------------------------------------
SS_2.dropna(inplace=True)
SS_2.head()
```

```python
# Step-5: Recode groups in the 'GENDER' column
# --------------------------------------------------
SS_2['GENDER'] = SS_2['GENDER'].map({0: 'Male', 1: 'Female'})
SS_2.head()
```

```python
# Step-6: Recode groups in the 'CL_YEAR' column
# --------------------------------------------------
SS_2['CL_YEAR'] = SS_2['CL_YEAR'].map({1: 'Freshman', 2: 'Sophomore', 3: 'Junior', 4: 'Senior'})
SS_2.head()
```

```python
# Step-7: Exporting file as CSV to the current working directory
# --------------------------------------------------
output_file_name = 'Sleep_Study.csv'
SS_2.to_csv(output_file_name, index=False)

print(f"Dataset successfully cleaned and saved to {output_file_name}")
```
