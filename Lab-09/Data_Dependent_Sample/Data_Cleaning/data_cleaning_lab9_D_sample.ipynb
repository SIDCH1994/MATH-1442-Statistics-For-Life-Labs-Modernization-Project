# Dataset for Lab-9 Dependent Samples is Forest_Study_Raw.csv

Raw dataset file Forest_Study_Raw.csv is generated using the random normal and gamma distributions

```python
# Step-1: Importing the raw dataset.
# --------------------------------------------------
import pandas as pd
FS = pd.read_csv(r'../Data_Generation/Forest_Study_raw.csv')
FS
```

```python
# Step-2: Removing unwanted variables
# --------------------------------------------------
cols_to_remove = ['Pre_PM25_ug_m3', 'Post_PM25_ug_m3']

FS_1 = FS.drop(columns=cols_to_remove)
FS_1.head()
```

```python
# Step-3: Renaming the column names
# --------------------------------------------------
rename_map = {'SiteID': 'SITE_ID',
              'Region': 'REGION',
              'VegType': 'VEG_TYPE',
              'Treatment': 'TREATMENT',
              'Elevation_m': 'ELEV',
              'Slope_deg': 'SLOPE',
              'CanopyCover_pct': 'CANOPY_COVER',
              'Pre_Biomass_Mg_ha': 'PRE_BIOMASS',
              'Post_Biomass_Mg_ha': 'POST_BIOMASS',
              'Pre_SoilMoisture_pct': 'PRE_SM',
              'Post_SoilMoisture_pct': 'POST_SM',
              'Pre_Temp_C': 'PRE_TEMP',
              'Post_Temp_C': 'POST_TEMP'}

FS_2 = FS_1.rename(columns=rename_map)

FS_2.head()
```

```python
# Step-4: Removing rows with NA or blank values.
# --------------------------------------------------
FS_2.dropna(inplace=True)
FS_2.head()
```

```python
# Step-5: Exporting file as CSV to the current working directory
# --------------------------------------------------
output_file_name = 'Forest_Resoration.csv'
FS_2.to_csv(output_file_name, index=False)

print(f"Dataset successfully cleaned and saved to {output_file_name}")
```
